**✅ Code to cycle colors automatically:**

```python
import matplotlib.pyplot as plt
import itertools

n = len(cat)
plt.figure(figsize=(24, 20))
for i in range(1, n):
    plt.subplot(3, 3, i)
    col = cat[i]
    keys = visa_df[col].value_counts().keys()
    values = visa_df[col].value_counts().values

    import itertools

    # Get the Set2 color palette (8 colors)
    base_colors = plt.colormaps['Set2'].colors
    
    # Create a color cycle to handle any number of categories
    color_cycle = itertools.cycle(base_colors)
    
    # Generate colors for all keys by cycling
    colors = [next(color_cycle) for _ in range(len(keys))]
    
    bars = plt.bar(keys, values, color=colors)


    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xticks(rotation=45)

    # adding count lables on top of bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, str(height),
                ha='center', va='bottom', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import itertools

n = len(cat)
plt.figure(figsize=(24, 20))

for i in range(1, n):
    plt.subplot(3, 3, i)
    col = cat[i]
    keys = visa_df[col].value_counts().keys()
    values = visa_df[col].value_counts().values

    # Prepare a cycling color list
    base_colors = plt.colormaps['Set2'].colors
    color_cycle = itertools.cycle(base_colors)
    colors = [next(color_cycle) for _ in range(len(keys))]

    # Plot and label each bar with a legend
    bars = []
    for key, value, color in zip(keys, values, colors):
        bar = plt.bar(key, value, color=color, label=key)
        bars.append(bar)

    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xticks(rotation=45)

    # Add value labels on top of bars
    for bar_group in bars:
        for bar in bar_group:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width() / 2, height, str(height),
                     ha='center', va='bottom', fontsize=12)

    plt.legend(title=col, fontsize=10, title_fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import itertools

n = len(cat)
plt.figure(figsize=(24, 20))

for i in range(1, n):
    plt.subplot(3, 3, i)
    col = cat[i]

    value_counts = visa_df[col].value_counts()
    percent_values = visa_df[col].value_counts(normalize=True) * 100  # Convert to percentage

    keys = value_counts.keys()
    values = value_counts.values
    percents = percent_values.values

    # Prepare cycling colors
    base_colors = plt.colormaps['Set2'].colors
    color_cycle = itertools.cycle(base_colors)
    colors = [next(color_cycle) for _ in range(len(keys))]

    bars = []
    for key, value, color in zip(keys, values, colors):
        bar = plt.bar(key, value, color=color, label=key)
        bars.append(bar)

    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xticks(rotation=45)

    # Add count + percentage labels
    for idx, bar_group in enumerate(bars):
        for bar in bar_group:
            height = bar.get_height()
            percentage = percents[idx]
            plt.text(bar.get_x() + bar.get_width() / 2,
                     height,
                     f'{int(height)}\n({percentage:.1f}%)',
                     ha='center', va='bottom', fontsize=11)

    plt.legend(title=col, fontsize=10, title_fontsize=12)

plt.tight_layout()
plt.show()
```
