```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data = {'prevailing_wage': [21077,20575]}
df = pd.DataFrame(data)
print("Original Data:\n", df)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))  # Default is [0,1], but can be changed (e.g., [-1,1])
```

```python
# Reshape to 2D array (required by sklearn)
wage_values = df['prevailing_wage'].values.reshape(-1, 1)

# Fit & transform in one step
scaled_wages = scaler.fit_transform(wage_values)

# Alternative (fit first, then transform):
# scaler.fit(wage_values)  # Compute min/max
# scaled_wages = scaler.transform(wage_values)  # Apply scaling

# Add scaled values back to DataFrame
df['wage_minmax'] = scaled_wages
print("\nScaled Data:\n", df)
```

```python
# Reshape to 2D array (required by sklearn)
wage_values = df['prevailing_wage'].values.reshape(-1, 1)

# Fit & transform in one step
scaled_wages = scaler.fit_transform(wage_values)

# Alternative (fit first, then transform):
# scaler.fit(wage_values)  # Compute min/max
# scaled_wages = scaler.transform(wage_values)  # Apply scaling

# Add scaled values back to DataFrame
df['wage_minmax'] = scaled_wages
print("\nScaled Data:\n", df)
```

```python
import joblib  # More efficient than pickle for large objects

# Save the scaler
joblib.dump(scaler, 'minmax_scaler.pkl')  

# (Alternative: Using pickle)
# import pickle
# with open('minmax_scaler.pkl', 'wb') as f:
#     pickle.dump(scaler, f)
```

```python
# Load the saved scaler
loaded_scaler = joblib.load('minmax_scaler.pkl')

# New data to transform
new_data = pd.DataFrame({'prevailing_wage': [21077,20575]})

# Apply scaling (must reshape like before)
new_wages = new_data['prevailing_wage'].values.reshape(-1, 1)
new_scaled = loaded_scaler.transform(new_wages)

new_data['wage_minmax'] = new_scaled
print("\nNew Scaled Data:\n", new_data)
```
