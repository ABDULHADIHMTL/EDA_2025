**LableEncoder  vs OneHotEncoder**

- Generally we dont want a relation between input columns

- we need both input columns are not correlated each othe 

- we need both input columns are **independent(X,Y)** each other

- we need both input columns are **perpendicular** each other

- we need both input columns are **orthogonal** each other 

- we need both input columns are **90 degrees phase shift** each other 

- OneHotEncoder will achieve above

- But OnehotEncode creates new columns

   - columns are increase means dimensions are increase

   - More dimension data need more processing time

   - More dimension data need more infra

   - This is called Curse of **Dimensionality**
